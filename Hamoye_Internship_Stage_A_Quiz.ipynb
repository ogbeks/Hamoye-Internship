### **Data Cleaning of missing data**
Since, we are aware of missing data only on the fuel_unit column, we need to fill the missing data by using the the **most common** values since we are not certain of the impact of the other corresponding column values. 
Thus, a strategy is fill by the most common is getting the mode of the fuel_unit and then fill the missing column value with that value.

*Note, one may discard the missing data if they are sure it won't affect the data analysis.* 


```
categorical_features=df.select_dtypes(include=np.object).columns.to_list()
print(categorical_features)
```

```
numerical_features=df.select_dtypes(exclude=np.object).columns.to_list()
print(numerical_features)
```

```
#Determine the frequency of the values in the fuel_unit column.
df.groupby('fuel_unit')['fuel_unit'].count()
```

```
#Filling the missing data on the fuel_unit by mcf and setting inplace to True to permanent the changes
df['fuel_unit'].fillna(value='mcf', inplace=True)
```

```
#check if missing values have been filled
df.isnull().sum()
```

#### **Checking for Outliers**
It is important to visualized data to determine outliers that may affect data analysis. This can be done using various ways such as using the python visualization libraries (Seaborn, matplotlib) methods such as scatterplot, boxplot and other mathematical functions such as z-score, IQR-score. Check https://towardsdatascience.com/ways-to-detect-and-remove-the-outliers-404d16608dba for more information to understand the outliers and how to remove them. The method that will be employ to ascertain the outliers and remove if necessary is the boxenplot from seaborn

```
#Plotting all the dataset to check for outliers
plt.figure(figsize=(20,4))
sns.boxenplot(data=df)
```


I observed that the dataset comparison of all isn't accurate as the scale takes recognition of the lower values features thus, we drop the lower values after ascertain that the outliers can be ignored(i.e false)


```
plt.figure(figsize=(20,4))
sns.boxenplot(data=df.drop(columns=['fuel_qty_burned', 'report_year', 'fuel_cost_per_unit_delivered', 'utility_id_ferc1','fuel_mmbtu_per_unit']))
```

```
#Checking the fuel_cost_per_mmbtu boxplot to ascertain the outliers
plt.figure(figsize=(13,4))
sns.boxplot(x=df['fuel_cost_per_mmbtu'])
```

```
#Droping the outlier of the fuel_cost_per_mmbtu for too far from the 100% quartile range
df[df['fuel_cost_per_mmbtu']>300000] #Get the index number for the outlier to be dropped
df.drop([28256], inplace=True)
```
