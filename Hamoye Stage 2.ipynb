```python
#droping the date and light column
df.columns
#df.drop(df[['date','lights']], axis=1, inplace=True)
df.head()
```

```python
#Importing the library to normalize the dataset
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

```

```python
#Normalizing the datasets using the MinMaxScaler from Sklearn
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
features_df =normalised_df.drop(columns=['Appliances'])
target_col = normalised_df['Appliances']
```

```python
#Importing the library that split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
#Spliting the dataset into training and testing set with a ratio 70:30 and a randomizing factor of 42
X_train, X_test, y_train, y_test = train_test_split(features_df, target_col, test_size=0.3, random_state=42)
```

```python
#Importing the method for linearRegression
from sklearn.linear_model import LinearRegression

```

```python
#Performing the multi-linear regression of the features and the targets using the training set
regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_predict = regressor.predict(X_test)
```

```python
#Importing the metrics method from the Sklearn library
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
#Calculating the Mean absolute error
mae=mean_absolute_error(y_test, y_predict)
print('Mean Absolute error: ',round(mae,4))
```

```python
#Calculating the R - squared
r2_score_value = r2_score(y_test, y_predict)
print('R-squared score: ', round(r2_score_value,2))
```

```python
#Calculating the Residual sum of squares
rss = np.sum(np.square(y_test - y_predict))
print('Residual sum of squares: ', round(rss, 2))
```

```python
#Calculating the  Mean Squared Error
mse = mean_squared_error(y_test, y_predict)
print('Mean Squared Error: ', round(mse,3))
```

```python
#Calculating the Root Mean Squared Error
rmse = np.sqrt(mse)
print('Root Mean Squared Error: ', round(rmse,3))
```

## Regularization of the model using Lasso and Ridge


```python
#Importing the Lasso and Ridge linear model regularization.
from sklearn.linear_model import Lasso, Ridge
```

```python
#Performing the Lasso and Ridge regularization of the model
lasso_reg =Lasso(alpha=0.001)
lasso_reg.fit(X_train, y_train)

ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(X_train, y_train)
```

```python
#Comparing the effect of the regularizations

def get_weight_df(model, features, col_name):
    weights= pd.Series(model.coef_, features.columns).sort_values()
    #print(model.coef_)
    #print(weights)
    weights_df = pd.DataFrame(weights).reset_index()
    #print(weights_df.columns)
    weights_df.columns=['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
#getting the weight of the model co-efficient of the multi-linear regression analysis
linear_model_weights = get_weight_df(regressor, X_train, 'Linear_Model_Weight')
print(linear_model_weights)
```

```python
#Getting the weight of the co-efficient of the Ridge regularization model
ridge_weights_df = get_weight_df(ridge_reg, X_train, 'Ridge_Weight')
print(ridge_weights_df)
```

```python
#Getting the weight of the co-efficient of the Lasso regularization model
lasso_weights_df = get_weight_df(lasso_reg, X_train, 'Lasso_weight')
print(lasso_weights_df)
```

```python
#Merging the dataframe of the linear_model weight, ridge_model weight and the Lasso_model weight
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
print(final_weights)
```
